```python
# Function to generate orthogonal matrices
def generate_orthogonal_matrix(key):
    np.random.seed(key)  # Seed the random generator
    random_matrix = np.random.rand(4, 4)

    # Gram-Schmidt process to create an orthogonal matrix
    def gram_schmidt(A):
        Q, _ = np.linalg.qr(A)
        return Q

    orthogonal_matrix = gram_schmidt(random_matrix)
    return orthogonal_matrix.flatten().reshape(16, 1)
```

```python

# Generate cancelable templates and prepare data for classification
images, labels_arr = [], []

for item in data_feature_label.values():
    vector_lx768 = np.array(item["left"][0]).reshape(1, 768)
    vector_rx768 = np.array(item["right"][0]).reshape(1, 768)
    key = int(item["id"])

    # Generate cancelable matrix
    orthogonal_matrix = generate_orthogonal_matrix(key)
    left_projection_matrix = np.dot(orthogonal_matrix, vector_lx768)
    right_projection_matrix = np.dot(orthogonal_matrix, vector_rx768)
    cancelable_matrix = np.multiply(left_projection_matrix, right_projection_matrix)

    images.append(cancelable_matrix)
    labels_arr.append(item["id"])
```

```python
# Prepare data for K-Fold Cross Validation
all_images = np.array(images).reshape(len(images), 16, 768, 1)
y_data_cat = to_categorical(labels_arr, num_classes=225)
```

# Define the CNN model

```python

def create_cnn_model():
    model = Sequential([
        layers.Flatten(input_shape=(16, 768, 1)),
        layers.Dense(500, activation='relu'),
        #layers.Dense(1024, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(225, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

# K-Fold Validation


```python


kf = KFold(n_splits=10, shuffle=True, random_state=42)
fold_no = 1
acc_per_fold, loss_per_fold = [], []

# Train and evaluate the model for each fold
for train_index, val_index in kf.split(all_images):
    X_train, X_val = all_images[train_index], all_images[val_index]
    y_train, y_val = y_data_cat[train_index], y_data_cat[val_index]

    model = create_cnn_model()
    print(f"Training for fold {fold_no} ...")
    history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_val, y_val), verbose=1)

    # Evaluate the model
    scores = model.evaluate(X_val, y_val, verbose=0)
    print(f"Score for fold {fold_no}: Loss = {scores[0]}, Accuracy = {scores[1]*100}%")
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])
    fold_no += 1

# Summary of results
print("\nAverage scores for all folds:")
print(f"> Accuracy: {np.mean(acc_per_fold)} (+- {np.std(acc_per_fold)})")
print(f"> Loss: {np.mean(loss_per_fold)}")
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(all_images, y_data_cat, test_size=0.2, random_state=42)

# Create and train the model
model = create_cnn_model()
print("Training the model...")
history = model.fit(X_train, y_train, epochs=12, batch_size=32, validation_data=(X_val, y_val), verbose=1)

# Evaluate the model
scores = model.evaluate(X_val, y_val, verbose=0)
print(f"\nValidation Results:")
print(f"Loss = {scores[0]}")
print(f"Accuracy = {scores[1] * 100}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Function to compute similarity scores (Euclidean distance in this case)
def compute_similarity(vector1, vector2):
    # Convert lists to NumPy arrays for computation
    vector1 = np.array(vector1)
    vector2 = np.array(vector2)
    return np.linalg.norm(vector1 - vector2)

# Generate similarity scores for genuine and impostor pairs
genuine_scores = []
impostor_scores = []

# Assume `data_feature_label` has the feature vectors
for key, item in data_feature_label.items():
    print(key)
    for other_key, other_item in data_feature_label.items():
        if key != other_key:
            if item["id"] == other_item["id"]:  # Genuine pair
                score = compute_similarity(item["left"][0], other_item["left"][0])
                genuine_scores.append(score)
            else:  # Impostor pair
                score = compute_similarity(item["left"][0], other_item["left"][0])
                impostor_scores.append(score)

# FAR and FRR calculation
thresholds = np.linspace(min(min(genuine_scores), min(impostor_scores)),
                         max(max(genuine_scores), max(impostor_scores)), 100)
far_values = []  # False Acceptance Rate
frr_values = []  # False Rejection Rate

for threshold in thresholds:
    # FAR: Proportion of impostor scores below the threshold
    far = sum(1 for score in impostor_scores if score <= threshold) / len(impostor_scores)
    # FRR: Proportion of genuine scores above the threshold
    frr = sum(1 for score in genuine_scores if score > threshold) / len(genuine_scores)
    far_values.append(far)
    frr_values.append(frr)

# Plot FAR and FRR
plt.figure(figsize=(10, 6))
plt.plot(thresholds, far_values, label='FAR (False Acceptance Rate)', color='blue')
plt.plot(thresholds, frr_values, label='FRR (False Rejection Rate)', color='red')
plt.xlabel('Threshold')
plt.ylabel('Rate')
plt.title('FAR and FRR Curve')
plt.legend()
plt.grid(True)
plt.show()

# Determine Equal Error Rate (EER)
differences = np.abs(np.array(far_values) - np.array(frr_values))
eer_index = np.argmin(differences)
eer_threshold = thresholds[eer_index]
eer = (far_values[eer_index] + frr_values[eer_index]) / 2

print(f"Equal Error Rate (EER): {eer * 100:.2f}% at threshold: {eer_threshold:.4f}")
```
